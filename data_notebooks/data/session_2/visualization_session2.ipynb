### -Data visualization is the visual presentation of data or information. 
### -The goal of data visualization is to communicate data or information clearly and effectively to readers. 
### -Typically, data is visualized in the form of a chart, infographic, diagram or map.

## Data visualization is used for:

- Making data engaging and easily digestible
- Identifying trends and outliers within a set of data
- Telling a story found within the data
- Reinforcing an argument or opinion
- Highlighting the important parts of a set of data

## Python gives us many ways to visualize our results and tell a story of our analysis

```python

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
movies_df['Duration'].plot()
```

## This doesnt give lot of infromation

Lets try it with a histogram

```python
movies_df['Duration'].plot.hist()
```

## Lets give this plot a title

```python
movies_df['Duration'].plot.hist(title="Frequency of movies by their duration")
plt.xlabel("Duration in minutes")
```

## [Q1] Let's try to plot a histogram of IMDB Score

```python
# Please complete the rest of this cell
movies_df['IMDB Score'].plot.hist(title="Film ratings by their IMDB score")
plt.xlabel("Scores")
```

```python
movies_df.plot.scatter(x='Duration', y='IMDB Score', title='Relation of movie ratings and their duration')
```

## Scatter Plots are often used to measure two variables on a single plot

```python
movies_df.plot.scatter(x='Budget', y='Gross Earnings', title='Relation of movie gross earnings to their budget')
```

## In order to understand visualization better lets remove some outliers
## Filter dataframe to only consider movie with budget < 3000000

```python
# Please complete the rest of this cell
filtered_df = movies_df[movies_df['Budget'] < 3000000]
filtered_df.plot.scatter(x='Budget', y='Gross Earnings', title='Relation of movie gross earnings to their budget')
```

## Multi line plots are also a good way to show two or more variables in a single plot
## For example,  lets visualize budget and gross earnings of movies

```python
movies_df[['Budget', 'Gross Earnings']].plot.line()
```

## Bringing it all together

## Lets plot a bar graph of top 10 movie directors

```python
# Step 1: Find top 10 movie directors
top_10_directors = movies_df.groupby('Director')['Title'].count().sort_values(ascending=False).head(10)

# Step 2: Plot them in graph
top_10_directors.plot(kind='bar')
```
